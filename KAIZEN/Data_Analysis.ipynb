### Official Grading System According to WAEC:

| Grade | Scale                | Grade Description |
|-------|----------------------|-------------------|
| A1    | 75.00 - 100.00       | Excellent         |
| B2    | 70.00 - 74.00        | Very Good         |
| B3    | 65.00 - 69.00        | Good              |
| C4    | 60.00 - 64.00        | Credit            |
| C5    | 55.00 - 59.00        | Credit            |
| C6    | 50.00 - 54.00        | Credit            |
| D7    | 45.00 - 49.00        | Pass              |
| E8    | 40.00 - 44.00        | Pass              |
| F9    | 0.00 - 39.00         | Fail              |


```python
# Custom function for rounding based on the rule
def round_custom(x):
    if x % 1 >= 0.5:
        return np.floor(x) + 1  # Round up
    else:
        return np.floor(x)       # Round down. remains the same

# List of columns to apply the rounding logic to
score_columns = ['SS1_avg_attendance', 'SS2_avg_attendance', 'SS3_avg_attendance', 'SS1_First_Term', 'SS1_Second_Term', 'SS1_Third_Term', 
           'SS2_First_Term', 'SS2_Second_Term', 'SS2_Third_Term', 
           'SS3_First_Term', 'SS3_Second_Term']

# Apply the custom rounding function to each column
df[score_columns] = df[score_columns].applymap(round_custom)

# Display the first few rows to verify the results
df.head()
```

```python
# creating new variables

# SS1 average (across 3 terms)
df['SS1_avg'] = df[['SS1_First_Term', 'SS1_Second_Term', 'SS1_Third_Term']].mean(axis=1).round().astype(int)

# SS2 average (across 3 terms)
df['SS2_avg'] = df[['SS2_First_Term', 'SS2_Second_Term', 'SS2_Third_Term']].mean(axis=1).round().astype(int)

# SS3 average (across 2 terms for SS3)
df['SS3_avg'] = df[['SS3_First_Term', 'SS3_Second_Term']].mean(axis=1).round().astype(int)

# Display the dataframe 
df.head()
```

```python
def classify_grade(score):
    if 75 <= score <= 100:
        return 'A1'
    elif 70 <= score <= 74:
        return 'B2'
    elif 65 <= score <= 69:
        return 'B3'
    elif 60 <= score <= 64:
        return 'C4'
    elif 55 <= score <= 59:
        return 'C5'
    elif 50 <= score <= 54:
        return 'C6'
    elif 45 <= score <= 49:
        return 'D7'
    elif 40 <= score <= 44:
        return 'E8'
    elif 0 <= score <= 39:
        return 'F9'

# Creating new grade columns based on the average scores
df['SS1_avg_grade'] = df['SS1_avg'].apply(classify_grade)
df['SS2_avg_grade'] = df['SS2_avg'].apply(classify_grade)
df['SS3_avg_grade'] = df['SS3_avg'].apply(classify_grade)

# Display the updated DataFrame with the new grade columns
df.head()
```

### Feature Distribution
We will look into the distribution of the features and gain some insight into this dataset and how the features might be treated.

```python
# Separate the numerical and categorical columns
numerical_columns = df.select_dtypes(include=[np.number]).columns.tolist()
categorical_columns = df.select_dtypes(exclude=[np.number]).columns.tolist()

# Plot the distribution of numerical features
plt.figure(figsize=(15, 10))
for i, feature in enumerate(numerical_columns):
    plt.subplot((len(numerical_columns) // 3) + 1, 3, i+1)  # Adjusted grid size dynamically
    sns.histplot(df[feature], kde=True)
    plt.title(feature)
    plt.xlabel('')
plt.tight_layout()
plt.savefig('numerical_columns.png') 
plt.show()
```

1. **Attendance Across SS1, SS2, and SS3**:
   - Seems to follow a relatively normal distribution. Most students tend to have attendance scores between 85 and 90. 
    
2. **Average Daily Study Hours**:
   - The uniform spread, indicates that students have varied study habits, with no strong preference for a particular number of daily study hours. However, there seems to be a slight peak around 3-4 hours per day.

3. **Term Scores for SS1, SS2, and SS3**:
   - The scores tend to cluster around the 65-75 range, indicating that most students perform at a "Good" or "Very Good" level based on WAEC grading.

4. **Yearly Averages (SS1_avg, SS2_avg, SS3_avg)**:
   - Shows that the majority of students have average scores between 65 and 75, which again points to a solid performance, aligning with "Good" and "Very Good" WAEC grades.

5. **No Extreme Outliers**: 
   - There are no strong outliers in terms of extremely high or low performance. The bulk of the scores are concentrated within a normal range.


```python
# Plot the distribution of categorical features
plt.figure(figsize=(12, 10))  # Increased figure size for better spacing
for i, feature in enumerate(categorical_columns):
    plt.subplot((len(categorical_columns) // 3) + 1, 3, i+1)  # Adjusted grid size dynamically
    sns.countplot(data=df, x=feature)
    plt.title(feature)
    plt.xlabel('')
plt.tight_layout()
plt.savefig('categorical_columns.png') 
plt.show()
```

The categorical data distributions depicted in the histograms provide insightful trends about various factors. Parents with secondary school education or NCE (Nigeria Certificate in Education) are the most represented, followed by those with MSc, HND, and Primary school education. This could indicate that a majority of the students come from households with moderate to advanced levels of parental education.  
The distribution of socioeconomic status is relatively balanced between high, middle, and low categories.  
Most students also come from families where parents are married, followed by single-parent households and a smaller number of divorced parents.

### Feature Correlation
Creating visualizations to understand the relationships between the variables and to investigate how features relate to each other, while identifying potential multicollinearity among the features.

```python
# Exclude non-numeric columns
numerical_columns = df.select_dtypes(include=[np.number])

# Calculate the correlation matrix
correlation_matrix = numerical_columns.corr()

plt.figure(figsize=(12, 10))
sns.set(style="whitegrid")

heatmap = sns.heatmap(
    correlation_matrix,
    annot=True, 
    fmt=".3f",  
    cmap='RdPu',
    vmin= -1  
)

plt.title('Enhanced Correlation Matrix', fontsize=20)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig('Enhanced_Correlation_Matrix.png')
plt.show()
```

```python
#Identify highly correlated pairs
high_corr = correlation_matrix.unstack().sort_values(kind="quicksort", ascending=False)
print(high_corr[(high_corr > 0.2) & (high_corr < 1)])  
```

### General Insights  
Consistency Across Terms: This could suggest that students who grasp the material well in one term are likely to continue performing well in subsequent terms.  

Predictive Value: The correlations suggest that early performance (first term scores) can be a good predictor of overall averages, which may be useful for educators in identifying students who may need additional support early in the academic year.  

Attendance Impact: The correlation between attendance and performance, although weaker, highlights the potential importance of attendance on academic success, suggesting that encouraging attendance could positively impact overall performance.
